```python
import pandas as pd
import time

# Web Scraping
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
namaproduk = []
keterangan = []
harga = []
terjual = []

driver = webdriver.Edge()

for page in range(1,4):
    url = f"https://www.itemku.com/game-key/browse/steam?page={page}" ##melakukan automasi pindah page
    driver.get(url)
    time.sleep(5)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")


    boxes = soup.find_all("a",{"class":"block h-full"})
    #Ambil code data item dari inspect element browser
    for box in boxes:
        elnama_produk = box.find("h2",{"class":"text-lg text-[#1B1B1B] font-semibold600 line-clamp-2"})
        el_keterangan = box.find("div",{"class":"mt-1 text-lg text-suva-grey line-clamp-2"})
        el_harga = box.find("div",{"class":"text-xl text-[#F46200] font-bold font-itemkuV2"})
        el_terjual = box.find("div",{"class":"text-base text-suva-grey"})

        try:
            namaproduk.append(elnama_produk.get_text())
        except:
            namaproduk.append(None)

        try:
            keterangan.append(el_keterangan.get_text())
        except:
            keterangan.append(None)

        try:
            harga.append(el_harga.get_text())
        except:
            harga.append(None)

        try:
            terjual.append(el_terjual.get_text())
        except:
            terjual.append(None)


driver.quit()


df = pd.DataFrame()

df["namaproduk"] = namaproduk
df["keterangan"] = keterangan
df["harga"] = harga
df["terjual"] = terjual

##Data Cleansing
df = df.drop_duplicates(subset=["namaproduk"]) 
df["harga"] = df["harga"].str.replace("Rp", "").str.replace(" ", "").str.replace(".", "").astype(int)
df["terjual"] = df["terjual"].str.replace("terjual", "").str.replace(" ", "").astype(int)
df["keterangan"] = df["keterangan"].str.replace("•", "")

df.to_csv("P0-CODA16-RMT-M1-MARDI_WICAKSANA.csv")
```
